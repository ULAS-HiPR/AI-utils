```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Synthetic NIR from RGB as a baseline

```python
img = cv2.imread('../assets/green-field.jpg')

rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
nir_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1,2,1)
plt.imshow(rgb_img)
plt.subplot(1,2,2)
plt.imshow(nir_img, cmap='gray')
```

# NDVI
Normalized Difference Vegetation Index (NDVI) is a simple graphical indicator that can be used to analyze remote sensing measurements, typically, but not necessarily, from a space platform, and assess whether the target being observed contains live green vegetation or not.

`NDVI = (NIR - RED) / (NIR + RED)`

```python
# calculate NDVI (Normalized Difference Vegetation Index) to [-1, 1]
ndvi = (nir_img.astype(float) - rgb_img[:,:,0].astype(float)) / (nir_img + rgb_img[:,:,0])
plt.figure()
plt.imshow(ndvi == 0, cmap='gray')
plt.colorbar()
plt.show()
```

# NDWI
Normalised Difference Water Index (NDWI) is a remote sensing derived index that is used globally to monitor changes in surface water. It is particularly useful for monitoring water bodies in forests and wetlands.

`NDWI = (NIR - SWIR) / (NIR + SWIR)`

```python
ndwi = (rgb_img[:,:,0] - nir_img) / (rgb_img[:,:,0] + nir_img)

plt.imshow(ndwi)
```
